# 1 - Data Preprocessing

Ham veriyi temizler, geçersiz URL'leri eler, kanonik URL üretir ve `data/processed/cleaned_dataset.csv` dosyasını yazar.

```python
import sys
from pathlib import Path

BASE_DIR = Path.cwd().resolve().parent if Path.cwd().name == 'src' else Path.cwd().resolve()
sys.path.insert(0, str(BASE_DIR / 'src'))

from url_guard.data import clean_dataset

raw_path = BASE_DIR / 'data' / 'raw' / 'malicious_phish.csv'
out_path = BASE_DIR / 'data' / 'processed' / 'cleaned_dataset.csv'
out_path.parent.mkdir(parents=True, exist_ok=True)

cleaned = clean_dataset(str(raw_path))
cleaned.to_csv(out_path, index=False)
print(cleaned.shape)
print(cleaned['label'].value_counts(normalize=True))
cleaned.head()
```
